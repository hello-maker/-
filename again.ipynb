```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import platform
from matplotlib import font_manager, rc 
import matplotlib
if platform.system() == 'Windows':
    path = "c:\Windows\Fonts\malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name() 
    rc('font', family=font_name)
```

```python
df = pd.read_excel("./new_data.xlsx", encoding = 'cp949')
df.head()
```

```python
df = df.drop(["환자수", "인구수", "환자수/인구수"], axis=1)
class_index=df["진료과목"].value_counts().index[0:47]
class_index[2]
df.head()
```

```python
df.head()
```

```python
region=list(df["시도"].value_counts().index)
```

```python
plt.figure(figsize=(12,400))
cnt = 0
for j in range(47):
    cnt +=1
    df_cnt = df.loc[df["진료과목"]==class_index[j]]
    plt.subplot(47,1,cnt)
    for i in range(17):
        plt.plot(df_cnt[df_cnt["시도"]==region[i]]["진료년도"],df_cnt[df_cnt["시도"]==region[i]]["백분율"],marker='o',linestyle='solid')
        plt.title(class_index[j])
        plt.legend(region, loc='upper right', bbox_to_anchor=(1.0, 1.0))
        plt.xlabel("연도")
        plt.ylabel("환자비율")
        plt.show
```
